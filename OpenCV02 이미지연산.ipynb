### OpenCV 이미지연산

```python
import cv2
```

```python
image = cv2.imread('cat.jpg')
```

```python
# 픽셀 수 및 이미지 크기 확인

print(image.shape)
print(image.size)
```

```python
# 이미지 Numpy 객체의 특정 픽셀을 가리킵니다.

px = image[100, 100]

# B, G, R 순서로 출력됩니다.
# (단, Gray Scale인 경우에는 B, G, R로 구분되지 않습니다.)

print(px)
```

```python
# R값만 출력하기

print(px[2])        #B G R 3개의 값중 R 인덱스는 2이다.
```

```python
# 특정 범위 픽셀 변경
import matplotlib.pyplot as plt
import time

start_time = time.time()

for i in range(0, 100):
    for j in range(0, 100):
        image[i, j] = [255, 255, 255]

print("---%s seconds ---" % (time.time() - start_time))

start_time = time.time()

image[0:100, 0:100] = [0, 0, 0]
print("---%s seconds ---" % (time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### ROI(Region of interesting; 관심있는 영역 추출)

```python
image = cv2.imread("cat.jpg")

# Numpy Slicing: R이 처리 가능
roi = image[200:350, 50:200]

# R이 단위로 이미지 복사하기
image[0:150, 0:150] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 픽셀별 색상 다루기
#모든픽셀에대한 R(빨간색 값)을 0으로 설정

image = cv2.imread("cat.jpg")
image[:,:, 2] = 0

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
