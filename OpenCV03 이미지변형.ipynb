### OpenCV 이미지 변형

#### 보간법을 사용하여 사이즈가 변할 때 픽셀 사이의 값을 조정할 수 있다.
#### cv2.resize(image, dsize, fx, fy, interpolation)
#### INTER_CUBIC: 사이즈를 크게 할 때 주로 사용.
#### INTER_AREA: 사이즈를 작게 할 때 주로 사용.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("cat.jpg")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

expand = cv2.resize(image, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(expand, cv2.COLOR_BGR2RGB))
plt.show()

shrink = cv2.resize(image, None, fx=0.8, fy=0.8, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(shrink, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 이미지 위치 변경

#### 변환 행렬 사용 특정한 좌표 a,b를 옮긴다.
                

```python
import numpy as np
```

```python
image = cv2.imread("cat.jpg")

# 행과 열 정보만 저장합니다.
height, width = image.shape[:2]

M = np.float32([[1, 0, 100], [0, 1, 10]])         #변환 행렬정의 a이동값(오른쪽으로50이동)
dst = cv2.warpAffine(image, M, (width, height))  # b는 y이동값 10만큼 이동 (아래로)

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 이미지 회전에는 회전변환을 위한 기본행렬 필요
#### cos -sin, sin cos 으로 이루어짐

```python
image = cv2.imread("cat.jpg")

# 행과 열 정보만 저장합니다.
height, width = image.shape[:2]

M = cv2.getRotationMatrix2D((width / 2, height / 2), 90, 0.5) # 너비와 높이의 중간값으로 센터를 지정
dst = cv2.warpAffine(image, M, (width, height))
print(M)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
